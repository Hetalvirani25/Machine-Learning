# Creating X and y variables

```python
X = data.drop(columns='Outcome',axis=1)
Y = data['Outcome']
```

# Splitting data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
```

```python
#splitting the data into training set and test set
print("Shape of X_train: ", X_train.shape)
print("Shape of y_train: ", Y_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_test: ", Y_test.shape)
```

# Running Classification models

## 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train,Y_train)
```

```python
logreg.pred = logreg.predict(X_test)
```

### Checking Accuracy

```python
#calculating the accuracy through Logistic Regression
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,logreg.pred)
```

### Classification Report and Confusion Matrix

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(Y_test,logreg.pred))
```

```python
print(classification_report(Y_test,logreg.pred))
```

## 2. SVM (Support Vector Machine)

```python
from sklearn import svm
```

```python
svm_model = svm.SVC(kernel='linear')
```

```python
svm_model.fit(X_train,Y_train)
```

```python
svm_model.pred = svm_model.predict(X_test)
```

```python
#calculating the accuracy through SVM
accuracy_score(Y_test,svm_model.pred)
```

```python
print(confusion_matrix(Y_test,svm_model.pred))
```

```python
print(classification_report(Y_test,svm_model.pred))
```

## 3. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model = RandomForestClassifier(n_estimators=200)
```

```python
rfc_model.fit(X_train,Y_train)
```

```python
rfc_model.pred = rfc_model.predict(X_test)
```

```python
#calculating the accuracy through random forest
accuracy_score(Y_test,rfc_model.pred)
```

```python
print(confusion_matrix(Y_test,rfc_model.pred))
```

```python
print(classification_report(Y_test,rfc_model.pred))
```

### Finding the variable which impact the most

#### Neural Network(Keras)

```python
from keras.models import Sequential
```

```python
model_3 = Sequential()
from keras.layers import Dense
model_3.add(Dense(16, activation='relu', input_dim=8))
model_3.add(Dense(16, activation='relu'))
model_3.add(Dense(1, activation='sigmoid'))
model_3.summary()
```

```python
model_3.compile(optimizer= 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'] )
```

```python
model_3.fit(X_train,Y_train,epochs=200)
```

```python
#accuracy through neural network 
score = model_3.evaluate(X_train,Y_train)
print('Accuracy on train dataset: %.2f' %(score[1]*100))
```

```python
# Load auto-mpg.data

data = pd.read_csv('auto-mpg.data',na_values='?',sep=r'\s+')
df=pd.DataFrame(data)

df.columns=['mpg', 'cylinders', 'displacement', 'horsepower','weight','acceleration','modelyear','origin','carname']

print(df)
# Hint: Use attribute information from website to determine column names
# Hint: Load with na_values = '?' and sep=r'\s+'
```

## Summarize values
What is the mean, std, min, max in each column?

```python
# Print mean for each column
df.mean()
```

```python
# where are the other columns? Check data types for each column
df.dtypes
```

```python
# Print full data statistics
df.describe()
```

We could be interested by these statistics in each of the origins. To get these, we first group values by origin, then ask for the description. We will only look at mpg for clarity

```python
# Print statistics by origin for mpg

df.groupby(by='origin').describe().mpg
```

## Find NaNs
How many NaNs in each column?

We can ask which entries are null, which produces a boolean array


```python
# Find which data entries are null

df.isnull()
```

Applying `sum()` to this boolean array will count the number of `True` values in each column

```python
# Find the sum of the boolean array
df.isnull().sum()
```

We can get complementary information from `info()`

```python
# Use info() on the dataset
df.info()
```

We can fill (replace) these missing values, for example with the minimum value in each column

```python
# Replace the missing values with the minimum value in each column

df.fillna(df.min())
# Print the statistics for the updated dataset
df.describe()
```

## Count unique values (a histogram)

We finish off, with our good friend the histogram

```python
# Print the number of unique values for the mpg column

df['mpg'].value_counts()
```
